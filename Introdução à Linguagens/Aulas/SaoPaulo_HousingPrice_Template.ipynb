### Estatísticas básicas com groupby

```python
# média dos valores de aluguel e total por bairro (district)

df.groupby('district')[['rent','total']].mean()
```

```python
# agrupamento a partir de agregação com estatística descritiva (média, desvio padrão, etc...)

df.groupby(['district','type']).agg(rent_mean=('rent','mean'),
                       rent_std=('rent','std'),                       
                       rext_max=('rent','max'))
```

## Funções de alta ordem

```python
# map
## criar uma coluna, com os nomes district com letras maisculas

df['district_upper'] = df['district'].map(lambda name: name.upper())
```

```python
df.head()
```

```python
# apply
## criar nova coluna que armazenará as taxas adicionais aplicadas ao aluguel (diferença entre os campos total e rent)

df["extra_charges"] = df.apply(lambda dado: dado["total"] - dado["rent"], axis=1)
```

```python
df.head()
```

```python
# replace

df["type"] = df["type"].replace(
    {"Studio e kitnet": "Studio", "Apartamento": "Apto", "Casa em condomínio": "Casa"}
)
```

```python
df['type'].unique()
```

```python
# substituindo "Rua" e "Avenida" nos endereços

# df["address"] = df["address"].str.replace({r'\bRua\b': 'R.', r'\bAvenida\b': 'Av.'}, regex=True)
df["address"] = df["address"].str.replace(r'\bRua\b', 'R.', regex=True).str.replace(r'\bAvenida\b', 'Av.', regex=True)
```

```python
df['address']
```

```python
# calcular a média do aluguel para cada tipo de imóvel
# armazenar uma nova coluna com esses dados

df['type_avg_rent'] = df.groupby('type')['rent'].transform('mean')
df.head()
```

```python
# criar um ranking de aluguel (maior para menor) por bairro

df['ranking_district_rent'] = df.groupby('district')['rent'].transform(lambda valor: valor.rank(ascending=False))
```

# Comparação de aluguel por Bairro

### Comparação dos valores de aluguel nos 10 bairros com mais imóveis no dataset

```python
df['district'].value_counts().head(10)
```

```python
# filtrar todo o dataframe com os 10 bairros com mais imóveis
maiores_bairros = df['district'].value_counts().head(10).index

df_maiores_bairros = df[df['district'].isin(maiores_bairros)]
```

```python
#gerar boxplot
sns.boxplot(data= df_maiores_bairros, x= 'district', y= 'rent', hue= 'district')
plt.title('Aluguel x Bairro')
plt.xlabel('Bairro')
plt.ylabel('Aluguel')
plt.xticks(rotation= 45) #rotacionar label do eixo x
plt.show()
```
