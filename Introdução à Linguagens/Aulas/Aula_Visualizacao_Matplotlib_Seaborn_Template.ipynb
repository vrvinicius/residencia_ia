
# Matplotlib e Seaborn

## Diferenças

### 1. Propósito Geral
- **Matplotlib**: Biblioteca de baixo nível que oferece controle total sobre os gráficos.
- **Seaborn**: Biblioteca de alto nível construída sobre o Matplotlib. Focada em simplificar a criação de gráficos estatísticos e análises exploratórias.

### 2. Estilo e Design
- **Matplotlib**: Gráficos básicos por padrão. Requer ajustes manuais para melhorar a estética.
- **Seaborn**: Possui estilos e paletas de cores padrão que tornam os gráficos mais atraentes.

### 3. Integração com Dados
- **Matplotlib**: Requer manipulação manual dos dados antes da visualização.
- **Seaborn**: Integra-se diretamente com DataFrames do Pandas.

### 4. Tipos de Gráficos
- **Matplotlib**: Permite criar qualquer tipo de gráfico com grande flexibilidade, mas requer mais código.
- **Seaborn**: Focado em gráficos estatísticos, como distribuições, boxplots e mapas de calor.



````python
# Importação das bibliotecas
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````


## 1. Gráfico de Dispersão (`scatterplot`)
Mostra a relação entre duas variáveis contínuas ou entre uma contínua e uma categórica, utilizando cores ou estilos para categorizações.


````python
# Dados de exemplo
data = pd.DataFrame({
    'x': np.random.rand(100),
    'y': np.random.rand(100),
    'categoria': np.random.choice(['A','B'],100)
})

data.head()
````

````python
# Matplotlib
## Dificuldade: separar por categorias
for cat in data['categoria'].unique():
    subset = data[data['categoria'] == cat]
    plt.scatter(subset['x'], subset['y'], label= cat)
plt.title('Gráfico de Dispersão')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.legend()
plt.show()
````

````python
# Seaborn
sns.scatterplot(data = data, x= 'x', y= 'y', hue= 'categoria', style= 'categoria')
plt.title('Gráfico de Dispersão')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.show()
````

## 5. Boxplot

- Usado para visualizar a distribuição de um conjunto de dados, destacando a mediana, os quartis (Q1 e Q3), o intervalo interquartil (IQR) e possíveis outliers.
- Ele fornece uma maneira rápida de identificar dispersão, simetria e valores extremos nos dados.
- O boxplot mostra:
  - **Mediana**: Linha central dentro do retângulo.
  - **Quartis**: Limites inferior e superior do retângulo (Q1 e Q3).
  - **Outliers**: Representados por pontos fora dos "bigodes".

### **Quartis no Boxplot**

1. **Primeiro Quartil (Q1 - 25%)**:
   - Representa o ponto abaixo do qual estão **25% dos dados**.
   - Indica o limite inferior do "caixa" do boxplot.

2. **Mediana (Q2 - 50%)**:
   - Representa o ponto central dos dados, onde **50% dos valores estão abaixo** e **50% estão acima**.
   - É representada pela **linha dentro do caixa** no boxplot.

3. **Terceiro Quartil (Q3 - 75%)**:
   - Representa o ponto abaixo do qual estão **75% dos dados**.
   - Indica o limite superior do "caixa" do boxplot.
   - Mostra onde os valores mais altos começam a se concentrar.

4. **Intervalo Interquartil (IQR)**:
   - É a diferença entre o terceiro e o primeiro quartil (\( IQR = Q3 - Q1 \)).
   - Ajuda a identificar a densidade e possíveis outliers.


### **Visualização no Boxplot**

```plaintext
                  Outliers
                  │
          --------┴-----------
                  │
                  │
        Limite Superior (Q3 + 1.5 × IQR)
                  │
     ┌───────────────────────────┐
Q3 ──┤                           │
     │          Caixa            │
     │                           │
 Q2 ─┼──────── Mediana ──────────┼
     │                           │
Q1 ──┤                           │
     └───────────────────────────┘
        Limite Inferior (Q1 - 1.5 × IQR)
                  │
          --------┴-----------
                  │
                  Outliers
```



````python
# Criando dados de exemplo
data = pd.DataFrame({
    'grupo': np.random.choice(['A', 'B', 'C', 'D'], 10000),
    'valores': np.random.normal(size= 10000)
})

data
````

````python
# Criando o boxplot com Seaborn
sns.boxplot(data= data, x= 'grupo', y= 'valores', hue= 'grupo')
plt.show()
````


### **Passos para Remover Outliers**

- A abordagem mais comum é baseada no **intervalo interquartil (IQR)**.


1. **Calcular o Intervalo Interquartil (IQR)**:
   - $( \text{IQR} = Q3 - Q1 )$

2. **Definir Limites para Outliers**:
   - Valores abaixo de $( Q1 - 1.5 \times \text{IQR} )$ ou acima de $( Q3 + 1.5 \times \text{IQR} )$

3. **Filtrar os Dados**:
   - Remova os valores que estão fora dos limites definidos.



````python
Q1 = data['valores'].quantile(0.25)
Q3 = data['valores'].quantile(0.75)

print(Q1)
print(Q3)
````

````python
def remove_outliers(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    return df[(df[coluna] >= limite_inferior) & (df[coluna] <= limite_superior)]
    
````

````python
data_sem_outliers = remove_outliers(data, 'valores')
````

````python
plt.figure(figsize= (10,6))
# boxplot com outlier
plt.subplot(1, 2, 1)
sns.boxplot(data= data, x= 'grupo', y= 'valores', hue= 'grupo')
plt.title('Com outlier')

# boxplot sem outlier
plt.subplot(1, 2, 2)
sns.boxplot(data= data_sem_outliers, x= 'grupo', y= 'valores', hue= 'grupo')
plt.title('Sem outlier')

plt.show()
````
